```python
date_cols = [
    'order_purchase_timestamp',
    'order_estimated_delivery_date',
    'review_creation_date', 'review_answer_timestamp']

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.info()
```

```python
categorical_columns = [
    'order_status',
    'customer_city',
    'customer_state',
    'seller_city',
    'seller_state',
    'payment_type',
    'product_category_name_english',
    'geolocation_state',
    'geolocation_city'
]

for col in categorical_columns:
    df[col] = df[col].astype('O')
```

```python
id_columns = [
    'order_id',
    'customer_id',
    'customer_unique_id',
    'product_id',
    'seller_id',
    'review_id'
]

for col in id_columns:
    df[col] = df[col].astype('O')
```

```python
int_columns = [
    'customer_zip_code_prefix',
    'order_item_id',
    'product_name_lenght',
    'product_description_lenght',
    'product_photos_qty',
    'seller_zip_code_prefix',
    'payment_sequential',
    'payment_installments',
    'review_score',
    'delivery_days'
]

for col in int_columns:
    df[col] = pd.to_numeric(df[col], downcast='integer')
```

```python
float_columns = [
    'price',
    'freight_value',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm',
    'payment_value',
    'geolocation_lat',
    'geolocation_lng'
]

for col in float_columns:
    df[col] = pd.to_numeric(df[col], downcast='float')
```

```python
df['delivery_days'] = df['delivery_days'].fillna(0).astype('int64')
```
